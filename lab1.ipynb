## Task 1

```python
# Copyright 2022 Ibrahem Mouhamad
#
#

frequency = 10                      # frequency of sine wave
overSampleRate = 30                 # oversmpling rate
fs = overSampleRate * frequency     # sampling frequency
phase = 1/3*math.pi                 # desired pahse shift in radians
nCyl = 5                            # to generate five cycles of sine waves

t = np.arange(0, nCyl*1/frequency-1/fs, 1/fs)

y = [math.sin(2*math.pi*frequency*ti+phase) for ti in t]
```

```python
# Copyright 2022 Ibrahem Mouhamad
#
#

plt.plot(t, y,)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Sine wave f=10 Hz')
plt.show()
```

## Task 2

```python
# Copyright 2022 Ibrahem Mouhamad
#
#

def sgn(x):
    if x == 0:
        return 0
    else:
        return -1 if x < 0 else 1

frequency = 10                      # frequency of sine wave
dutyCycle = 50                     # oversmpling rate
fs = dutyCycle * frequency     # sampling frequency
nCyl = 5                            # to generate five cycles of sine waves

t = np.arange(0, nCyl*1/frequency, 1/fs)

y = [sgn(math.sin(2*math.pi*frequency*ti)) for ti in t]
```

```python
# Copyright 2022 Ibrahem Mouhamad
#
#

plt.plot(t, y,)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Square wave f=10 Hz')
plt.show()
```

## Task 4

```python
# Copyright 2022 Ibrahem Mouhamad
#
#

def chrip_signal(t, f0, t1, f1, phase = 0):
    '''
    chrip_signal(t, f0, t1, f1, phase) generates samples of a linearly
    swept-frequency signal at the time instances defined in timebase array t.
    The instantaneous frequency at time 0 is f0 Hz. The instantaneous
    frequency f1 is achieved at time t1. The argument 'phase' is optional. It
    defines the initial phase of the signal defined in radians. By default
    phase=0 radian.
    '''
    t0 = t[0]
    T = t1 - t0
    k = (f1-f0)/T

    g = [math.cos(2*math.pi*(k/2*ti+f0)*ti+phase) for ti in t]
    return g
```

```python
# Copyright 2022 Ibrahem Mouhamad
#
#

fs = 500                            # sampling frequency
t1 = 1
t = np.arange(0, t1, 1/fs)           # time base - up to 1 second
f0 = 1                              # starting frequency of the chrip
f1 = fs/20                          # frequency of the chrip signal, at t1=1

g = chrip_signal(t, f0, t1, f1)
```

```python
# Copyright 2022 Ibrahem Mouhamad
#
#

import matplotlib.pyplot as plt

plt.plot(t, g)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Chirp Signal')
plt.show()
```
